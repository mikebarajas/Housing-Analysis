```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
```

```python
# Load in csv
filepath = os.path.join("Employment.csv")
employment_df = pd.read_csv(filepath)
filepath = os.path.join("Unemployment.csv")
unemployment_df = pd.read_csv(filepath)
filepath = os.path.join("Income.csv")
income_df = pd.read_csv(filepath)
filepath = os.path.join("Population.csv")
population_df = pd.read_csv(filepath)
filepath = os.path.join("Price.csv")
price_df = pd.read_csv(filepath)
filepath = os.path.join("Rent.csv")
rent_df = pd.read_csv(filepath)
```

```python
# Merge the Employment and Unemployment Data Frames
merge_df = pd.merge(employment_df, unemployment_df, on=["Year", "Period ID", "Period", "Area"])

# Select desired columns 

merge_df = merge_df.drop(merge_df.columns[[2,4, 6]], axis=1)
merge_df.head()

# Change column name of Area to City
merge_df=merge_df.rename(columns = {'Area':'City', 'Period ID': 'Month'})

# Slice out MSA out of the City Names
merge_df['City'],merge_df['trash'] = merge_df["City"].str.split(' MSA', 2).str[0:2].str
city_df = merge_df.drop('trash', axis=1)
city_df.head()
```

```python
g = city_df.groupby(['City', 'Year'])
g.describe()
```

```python
# https://www.youtube.com/watch?v=Wb2Tp35dZ-I&t=2s
for city, city_df in g:
    print(city)
    print(city_df)
```
